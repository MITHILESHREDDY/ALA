```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import style
```

```python
df_google = pd.read_csv("GOOG.csv")
df_google.head()
```

```python
df_google.info() # check the data types in google csv dataframe 
```

```python
df_google.dtypes # short - check the data types in google csv dataframe
```

```python
df_google['date'] = pd.to_datetime(df_google['date'], format='%Y-%m-%d')
df_google.info()
```

```python
style.use("ggplot")
df_google.plot(x="date", y="close", figsize=(16,8))
plt.show()
```

```python
closelist = list(df_google.close)
window_size=100
i=0
moving_avg=[]

while i<len(closelist)-window_size+1:
    w=closelist[i:i+window_size]
    avg=round(sum(w)/window_size,2)
    moving_avg.append(avg)
    i+=1

for i in range(0,99):
    moving_avg.insert(0,None)

df_google['movingavg']=moving_avg
df_google.plot(x='date',y='movingavg',figsize=(16,8))
plt.show()
```

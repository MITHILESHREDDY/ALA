```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from matplotlib import style
```

```python
df_google = pd.read_csv("GOOG.csv")
df_google.head()
```

```python
df_google.info() # check the data types in google csv dataframe 
```

```python
df_google.dtypes # short - check the data types in google csv dataframe
```

```python
df_google['date'] = pd.to_datetime(df_google['date'], format='%Y-%m-%d')
df_google.info()
```

```python
style.use("ggplot")
df_google.plot(x="date", y="close", figsize=(16,8))
plt.show()
```

```python
# Soomthing the plot by using rolling mean. This performs moving average i.e., calculating the average over a window of data points.
# Considering the size of the moving window as 100 to achieve the desired plot

df_google =df_google.set_index('date')
df_google['close'].rolling(100).mean().plot(figsize=(16,8)) 
```
